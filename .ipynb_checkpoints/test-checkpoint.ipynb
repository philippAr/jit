```python
%pylab inline
import numpy as np
from time import time
from matplotlib import pyplot as plt
```

```python
data = np.array([])
start = time()
while (input() == ''):
    data = np.append(data, time() - start)
data = data[1:]

name = input('Save as: ')
np.save('data/'+name , data)
np.savetxt('data/'+name , data , delimiter = ';')
```

```python
werk = input('Lade Datei:')
data = np.load('data/' + werk + '.npy')
```

```python
def computeBPM(data , steps , conversionfactor):
    proc = np.array([])
    for i in range(np.shape(data)[0] - steps):
        proc = np.append(proc , steps / (data[i + steps] - data[i]) * 60 / conversionfactor)
    return proc
```

```python
for j in [1,20,80,200]:
    glaettung = j
    conversion = 1
    bartype = 4 # after conversion
    #fromgl = 1
    #to = 4
    #BPM = computeBPM(data, fromgl , conversion)[:len(computeBPM(data, to , conversion))] - computeBPM(data, to , conversion)
    BPM = computeBPM(data, glaettung , conversion)
    mw = np.mean(BPM)
    std = np.std(BPM)
    print('Mittelwert: ' + str(np.round(mw,1)) + ' +/- ' + str(np.round(std,1)))
    print('Extrema: ' + str(np.round(np.min(BPM),1)) + ' -- ' + str(np.round(np.max(BPM),1)))

    mwr = np.array([])
    stdr = np.array([])
    for i in range(len(BPM)):
        mwr = np.append(mwr , mw )
        stdr = np.append(stdr , std )

    # als x-Achse: Zeit
    xaxis = data[:len(BPM)] / 60

    # als x-Achse: Takte
    xaxis = np.arange(len(BPM)) / bartype / conversion

    plt.figure(figsize = (24,6))
    plt.plot(xaxis, BPM , label = str(i))
    plt.plot(xaxis , mwr)
    plt.plot(xaxis , mwr + stdr)
    plt.plot(xaxis , mwr - stdr)
    plt.ylabel('BPM')
    plt.xlabel('Takte')
    plt.ylim(mw - 3 * std , mw + 2 * std )
    plt.xticks(np.arange(min(xaxis), max(xaxis)+1, 10.0))
    plt.savefig('pics/' + werk + '(' + str(j) + ').png' , dpi = 300)
    #plt.show()
```
