```python
%pylab inline
import numpy as np
from time import time
from matplotlib import pyplot as plt
```

```python
data = np.array([])
start = time()
while (input() == ''):
    data = np.append(data, time() - start)
data = data[1:]
```

```python
def computeBPM(data , steps , conversionfactor):
    proc = np.array([])
    for i in range(np.shape(data)[0] - steps):
        proc = np.append(proc , steps / (data[i + steps] - data[i]) * 60 / conversionfactor)
    return proc


glaettung = 5
BPM = computeBPM(data, glaettung , 1)
mw = np.mean(BPM)
std = np.std(BPM)
print('Mittelwert: ' + str(np.round(mw,1)) + ' +/- ' + str(np.round(std,1)))
print('Extrema: ' + str(np.round(np.min(BPM),1)) + ' -- ' + str(np.round(np.max(BPM),1)))

mwr = np.array([])
stdr = np.array([])
for i in range(len(BPM)):
    mwr = np.append(mwr , mw )
    stdr = np.append(stdr , std )
    
xaxis = data[:len(BPM)] / 60    

plt.plot(xaxis, BPM , label = str(i))
plt.plot(xaxis , mwr)
plt.plot(xaxis , mwr + stdr)
plt.plot(xaxis , mwr - stdr)
plt.ylabel('BPM')
plt.xlabel('Zeit [min]')
plt.show()
```
